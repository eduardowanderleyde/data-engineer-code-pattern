```python
import pandas as pd
import json
```

# 1.Loading data from a CSV file

### We will load the sales.csv file which contains sales information.

CSV - Comma-Sepated Values

```python
# Loading the CSV file
df_vendas = pd.read_csv('vendas.csv')

# Displaying the first lines
print("First 5 lines of the sales file:")
display(df_vendas.head())

# Information about the Dataframe
print("Information about the Dataframe:")
display(df_vendas.info())
```

# 2. Loading data from a JSON file

### Now we will load the cadastro.json file which contains information about customers.

```python
# Loading the JSON file
with open('cadastro.json', 'r', encoding='utf-8') as file:
    data_json = json.load(file)  # dict

print(type(data_json))

# Converting to DataFrame
df_clients = pd.DataFrame(data_json['clientes'])

# Displaying the first rows
print("First 5 rows of the registration file if available:")
display(df_clients.head())

# Information about the DataFrame
print("\nInformation about the DataFrame:")
display(df_clients.info())
```
